```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
from matplotlib.pyplot import *

psnr_scale_4 = [    25.53,   25.49,     25.38,   25.33,  25.93,      26.12,     25.93,     26.28]
no_of_parameters = [7,       100,      20,      0.002,   0.9,          0.05,      19,              0.005]
Name_Of_Models = ["A+",     "SelfExSR",   "RFL",  "SRCNN", "VDSR",  "LapSRN",  "DRCN",     "MCISIR"]
for k in range(8):
    loc_x = no_of_parameters[k]
    loc_y = psnr_scale_4[k]
    if loc_x == 0.005 and loc_y == 26.28:
        plt.semilogx(loc_x,loc_y,color='b',marker='.')
        plt.text(loc_x+0.01,loc_y,str(Name_Of_Models[k]),fontsize=12,color='b')
    else:
        plt.semilogx(loc_x,loc_y,color='r',marker='.')
        plt.text(loc_x+0.01,loc_y,str(Name_Of_Models[k]),fontsize=12)
    
plt.ylim([25,26.4])
plt.xlim(100,0.0018)
#plt.xlim(10,1000)
plt.title("PSNR Versus Running time (s)")
plt.ylabel('PSNR(dB)',fontsize=16)
plt.xlabel('slower  <--------       Running time (s)        ---------->        faster',fontsize=12)
plt.grid(which="both")
plt.show()
```
